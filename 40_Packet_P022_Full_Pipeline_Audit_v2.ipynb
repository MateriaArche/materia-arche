```python
# ── Pipeline Credibility Scoring ────────────────────────────────────
confirmed  = (packets['Status'] == 'Confirmed').sum()
promising  = (packets['Status'] == 'Promising').sum()
negative   = (packets['Status'] == 'Negative').sum()
total      = len(packets)
credibility = confirmed / total

print("=" * 70)
print("PIPELINE CREDIBILITY SCORE")
print("=" * 70)
print(f"  Confirmed : {confirmed}")
print(f"  Promising : {promising}")
print(f"  Negative  : {negative}")
print(f"  Total     : {total}")
print(f"  Credibility (Confirmed / Total): {credibility:.1%}")
print()

neg = packets[packets['Status'] == 'Negative']
print("NEGATIVE PACKETS AND IMPLICATIONS:")
print("-" * 70)
implications = {
    'P-004': 'Full-dataset ranking memorises training data -> corrected by P-005',
    'P-006': 'Panel lacked diversity -> corrected by P-007',
    'P-011': 'Multi-model agreement limited to 1/3 panel devices -> known limitation',
    'P-013': 'Leave-one-group-out CV near zero -> model does not generalise across labs',
    'P-015': 'Naive prediction intervals under-cover -> corrected by P-017',
    'P-016': 'Novel composition holdout near zero -> model does not extrapolate',
    'P-018': 'SHAP top-3 unstable across splits -> corrected by P-020',
}
for _, row in neg.iterrows():
    pid = row['Packet']
    impl = implications.get(pid, 'See notebook for details')
    print(f"  {pid}: {row['Title']}")
    print(f"         {impl}")
    print()

print("NEGATIVES ADDRESSED:")
print("  P-004 (shortlist bias)          -> P-005 (test-set-only ranking)")
print("  P-006 (MA-only panel)           -> P-007 (diversified panel)")
print("  P-015 (interval under-coverage) -> P-017 (conformal calibration)")
print("  P-018 (SHAP instability)        -> P-020 (6 consensus features)")
print()
print("NEGATIVES REMAINING (unresolved):")
print("  P-011 (partial multi-model consensus)")
print("  P-013 (publication / lab bias)")
print("  P-016 (no extrapolation to novel families)")
```

```python
# ── Reproduce Key Numbers ───────────────────────────────────────────
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import train_test_split
from scipy.stats import kendalltau

df = pd.read_csv('perovskite_stability_clean.csv')

FEATURES = [
    'Perovskite_band_gap', 'Pb', 'Sn', 'I', 'Br', 'Cl',
    'MA', 'FA', 'Cs',
    'first_Prvskt_annealing_temperature',
    'first_Prvskt_thermal_annealing_time',
    'Perovskite_thickness', 'Cell_area_measured',
    'JV_default_Voc', 'JV_default_Jsc', 'JV_default_FF',
]
TARGET = 'Stability_PCE_T80'

X = df[FEATURES].fillna(0)
y = np.log1p(df[TARGET].fillna(0))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

et = ExtraTreesRegressor(
    n_estimators=700,
    max_features='sqrt',
    min_samples_split=3,
    min_samples_leaf=1,
    bootstrap=False,
    random_state=42,
)
et.fit(X_train, y_train)
y_pred = et.predict(X_test)

tau, pval = kendalltau(y_test, y_pred)

# Panel device rankings
PANEL = [850, 1320, 119]
panel_names = {
    850:  'MA3Pb4I13 (MA-only)',
    1320: 'MA0.25FA0.75PbI2.77Br0.25 (MA-FA mixed)',
    119:  'FA0.83Cs0.17PbI2Br (FA-Cs triple)',
}

test_indices = X_test.index
preds_series = pd.Series(y_pred, index=test_indices)
ranks = preds_series.rank(ascending=False)
n_test = len(test_indices)

print("=" * 70)
print("REPRODUCED KEY NUMBERS  (seed=42, single split)")
print("=" * 70)
print(f"  Dataset: {len(df)} devices, {len(FEATURES)} features")
print(f"  Train/test: {len(X_train)} / {len(X_test)}")
print(f"  Test tau-b: {tau:.4f}  (p = {pval:.2e})")
print()
print("  Panel device rankings on test set:")
for dev_id in PANEL:
    if dev_id in test_indices:
        r = int(ranks.loc[dev_id])
        actual_t80 = df.loc[dev_id, TARGET]
        pred_t80 = np.expm1(preds_series.loc[dev_id])
        print(f"    Device {dev_id} ({panel_names[dev_id]})")
        print(f"      Rank: {r}/{n_test}  |  Actual T80: {actual_t80:.0f}h  |  Predicted T80: {pred_t80:.0f}h")
    else:
        print(f"    Device {dev_id}: in training set for this split")
print()
print(f"  Claim check: tau-b ~ 0.27 -> measured {tau:.4f}  {'PASS' if abs(tau - 0.27) < 0.05 else 'CHECK'}")
```
